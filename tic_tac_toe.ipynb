```python
def user_choice():
    c1 = ' '
    range1 = range(1,10)
    is_range = False
    while c1.isdigit()==False or is_range==False:
        c1 = input('choose 1-9: ')
        if c1.isdigit()==False:
            print('not digit')
            
        if c1.isdigit()==True:
            if int(c1) in range1:
                is_range == True
                return int(c1)
            else:
                pass
                print('not in range')
```
